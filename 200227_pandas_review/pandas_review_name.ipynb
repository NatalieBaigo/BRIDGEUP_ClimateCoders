#### Find the monthly mean

Remember!
1. "your data".mean(axis = **0**) - mean along the columns
1. "your data".mean(axis = **1**) - mean along the rows

```python
#we made a new copy of our dataset but this time the year isnt considered as a column but more as a title
data= data.set_index("Year")
data
```

```python
#here we found the monthly average for our data 
avg= data.mean(axis=0)
avg
```

#### Plot your result

```python
month= data.columns
print(avg)
print(month)
data.index
plt.plot(month,avg)
```
